```python
import pandas as pd
import numpy as np


# Encodding
```

```python
df = pd.read_csv('../data/Telco-Customer-Churn.csv')
df
```

```python
df.columns
```

```python
# Binary Encoding / Yes-No

binary_yn = [col for col in df.columns if df[col].dropna().nunique() == 2 and set(df[col].dropna().unique()).issubset({'yes', 'no'})]

df[binary_yn] = df[binary_yn].map({'yes', 'no'})
```
